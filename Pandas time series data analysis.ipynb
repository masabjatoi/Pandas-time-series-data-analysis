```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
import datetime as dt
```

```python
df = pd.read_csv("AAPL.csv")
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
```

```python
df["High"].resample("M").mean()
```

```python
df = df.sort_values(by = "Date")
```

```python
newdf = df.loc["2023-01-01" : "2023-5-30"]
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=newdf , label='High - Low', marker='o')
plt.title('Difference Between High and Low')
plt.xlabel('Date')
plt.ylabel('Difference')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid()
plt.show()
```

```python
Monthly_data = df["High"].resample("Q").mean()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(Monthly_data , label="High")
plt.title(' Monthly Avg of High Stocks ')
plt.xlabel('Date')
plt.ylabel('Ratio')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid()
plt.show()
```
